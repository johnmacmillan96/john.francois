## Using R with a Python [conda root] kernal

The Python [conda root] kernal utilizes built-in "magic" Jupyter commands denoted by a '%' for single line and '%%' for block. Using the %load_ext magic command, we can load an incredibly useful *rmagic* extension which can convert inline code into R language, pulling in the data from the notebook and pushing the output as Python.

The *rmagic* exension is part of the *rpy2.ipython* extension, which comes pre-installed with Jupyter. To start using R, load both extensions as follows:

````python
%load_ext rpy2.ipython
%load_ext rmagic
````

# How to push and pull data

The '%R' command, followed by in-line code, executed the line of code in an R environment, and the output is returned to the (Python) environment of this Jupyter Notebook. Lines of code can be seperated by semicolons.

The '%%R' command, followed by a block of code, executes the block of code in an R environment. By default, no values are returned to the (Python) environment of this Jupyter Notebook.

There are optional arguments that can procede the code. They are as follows (taken from https://ipython.org/ipython-doc/2/config/extensions/rmagic.html):

```
-i INPUT, --input INPUT
    Names of input variable from shell.user_ns to be assigned to R variables of the same names after 
    calling self.pyconverter. Multiple names can be passed separated only by commas with no whitespace.
    
-o OUTPUT, --output OUTPUT
 	Names of variables to be pushed from rpy2 to shell.user_ns after executing cell body and applying 
    self.R converter. Multiple names can be passed separated only by commas with no whitespace.

-w WIDTH, --width WIDTH
 	Width of png plotting device sent as an argument to png in R.

-h HEIGHT, --height HEIGHT
 	Height of png plotting device sent as an argument to png in R.

-d DATAFRAME, --dataframe DATAFRAME
 	Convert these objects to data.frames and return as structured arrays.

-u <{px,in,cm,mm}>, --units <{px,in,cm,mm}>
 	Units of png plotting device sent as an argument to png in R. One of [“px”, “in”, “cm”, “mm”].

-r RES, --res RES
 	Resolution of png plotting device sent as an argument to png in R. Defaults to 72 if units is one of 
    [“in”, “cm”, “mm”].

-p POINTSIZE, --pointsize POINTSIZE
 	Pointsize of png plotting device sent as an argument to png in R.
    
-b BG --b BG
    Background of png plotting device sent as an argument to png in R.
    
-n -- noreturn
    Force the magic to not return anything.
```

````python
%R library(popcycle)
````

````python
%%R
plot.time = function(db, pop) {
    table <- get.stat.table(db)
    population.list <- subset(table, pop == "prochloro")
    population.list.filtered <- population.list[(population.list$n_count < 80000),]
    population.list.filtered <- population.list.filtered[(population.list.filtered$n_count > 16000),]
    time.stat <- as.POSIXct(population.list.filtered$time, format = "%FT%T", tz = "GMT")
    pop.count <- population.list.filtered$n_count 
    plot(time.stat, pop.count)
}
````
